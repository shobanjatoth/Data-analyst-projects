```python
price=pd.read_csv('price_data (1).csv')
```

```python
price.head()
```

```python
price.columns
```

```python
 #id = client company identifier
 #price_date = reference date
# price_off_peak_var = price of energy for the 1st period (off peak)
# price_peak_var = price of energy for the 2nd period (peak)
#price_mid_peak_var = price of energy for the 3rd period (mid peak)
#price_off_peak_fix = price of power for the 1st period (off peak)
#price_peak_fix = price of power for the 2nd period (peak)
#price_mid_peak_fix = price of power for the 3rd period (mid peak)
```

```python
price.shape
```

```python
price.info()
```

```python
price.describe()
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(data=price, x='price_date', y='price_off_peak_var', label='Off-Peak Variable Price')
sns.lineplot(data=price, x='price_date', y='price_peak_var', label='Peak Variable Price')
sns.lineplot(data=price, x='price_date', y='price_mid_peak_var', label='Mid-Peak Variable Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Price Variations Over Time')
plt.legend()
plt.show()
```

# observation
